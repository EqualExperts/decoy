```python
import pandas as pd
```

```python
names = pd.DataFrame({'names': ['A', 'B', 'C', 'D', 'E']}).reset_index()
display(names)

addresses = pd.DataFrame({'addresses': ['z', 'y', 'x', 'w', 'v']}).reset_index()
display(addresses)
```

```python
names.merge(addresses, how='left', on='index')
```

```python
nrows = len(names) * len(addresses)
print(nrows)
```
